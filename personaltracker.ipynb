```python
# Personal Finance Tracker (Google Colab Version)

import pandas as pd
import matplotlib.pyplot as plt

# -----------------------------
# Sample Expense Data
# -----------------------------
expenses = {
    "Date": [
        "2025-06-01",
        "2025-06-02",
        "2025-06-03",
        "2025-06-05",
        "2025-06-07",
        "2025-06-10"
    ],
    "Category": [
        "Food",
        "Transport",
        "Shopping",
        "Bills",
        "Food",
        "Entertainment"
    ],
    "Amount": [
        250,
        120,
        1000,
        2000,
        300,
        500
    ]
}

df = pd.DataFrame(expenses)

print("Expense Records")
display(df)

# -----------------------------
# Total Expenses
# -----------------------------
total_expense = df["Amount"].sum()
print(f"\nTotal Expenses: ₹{total_expense}")

# -----------------------------
# Category-wise Spending
# -----------------------------
category_expense = df.groupby("Category")["Amount"].sum()

print("\nCategory-wise Expenses")
display(category_expense)

# -----------------------------
# Bar Chart
# -----------------------------
plt.figure(figsize=(8,5))
category_expense.plot(kind="bar")
plt.title("Expenses by Category")
plt.xlabel("Category")
plt.ylabel("Amount (₹)")
plt.grid(True)
plt.show()

# -----------------------------
# Pie Chart
# -----------------------------
plt.figure(figsize=(6,6))
category_expense.plot(
    kind="pie",
    autopct="%1.1f%%"
)
plt.title("Expense Distribution")
plt.ylabel("")
plt.show()

# -----------------------------
# Savings Prediction
# -----------------------------
monthly_income = 10000

predicted_savings = monthly_income - total_expense

print("\nMonthly Income: ₹", monthly_income)
print("Predicted Savings: ₹", predicted_savings)

if predicted_savings > 0:
    print("You are saving money.")
else:
    print("Expenses exceed income.")
```
